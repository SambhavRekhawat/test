## MEAN REVERSION PROCESS WITH DRIFT

Consider a CPI with a starting value of 1, drift rate of 5%, annualized volatility of 25%, long-term mean reversion rate of 5, reversion rate of 25% and a forecast horizon of 10 years. Let’s predict the CPI future value 10 years from today, based on those parameters and a stochastic process called mean-reversion with drift.

Since we wish to run a path-dependent simulation, we will use 100 time steps. Finally, we will use 50,000 iterations in the calculation.

Thus, S0 = 1, T = 10.0, μ = 0.05, σ = 0.25, Sm = 5, η = 0.25, N=100, M = 50,000

image.png

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from numpy import random as rn
from scipy import stats
import scipy.stats as si
import seaborn as sns
G = 1629562571
np.random.seed(G)
from datetime import datetime
# datetime object containing current date and time
now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
```

```python
S0 = 1
T = 10
μ = 0.05
σ = 0.25
Sm = 5
η = 0.25
M = 50000
N = 100
N
```

```python
ε = rn.randn(M,N)
S = S0*np.ones((M,N+1))
dt = T/N
dt
```

```python
start_time = datetime.now()
for i in range(0,N):
    S[:,i+1] = S[:,i]*(1 + η*(Sm*np.exp(μ*dt)-S[:,i])*dt + μ*dt + σ*ε[:,i]*np.sqrt(dt))
```

```python
plt.figure(figsize=(10,6))
fontsize=15
plt.title('Path-Dependent Monte Carlo Simulation - Mean-Reversion Process with Drift',fontsize=fontsize)
plt.xlabel('Years',fontsize=fontsize)
plt.ylabel('CPI prices (USD)',fontsize=fontsize)
plt.grid(axis='y')
a = [ rn.randint(0,M) for j in range(1,40)]
for runer in a:
    plt.plot(np.arange(0,T+dt,dt), S[runer], 'red')
```

```python
V = (S[:,-1])
print("\033[1m The estimate of the future value of the CPI is ${:.2f}".format(np.mean(V)))
print("\033[1m The accuracy of the estimate is ${:.2f}".format((np.std(V)/np.sqrt(M))))
```
